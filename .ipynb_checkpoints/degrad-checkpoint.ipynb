# generic bulk model
from mass balance + fick's (darling)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import trapz, simps

c_feed = 1/1000
z = 1
alpha = 0.5
T = 273.15 + 80
RT = 8.314*T
E_eq = 0.0
E = np.arange(-0.3,0.3,0.05)
n = 2

a_v = 4e10
L = 15e-6
x = np.arange(0, L+0.5e-6, 0.5e-6)
i_0 = 2e-4
F = 96485
D = 0.77e-5

def beta(i):
    return alpha*z*F/RT*(E[i]-E_eq)

## reaction, called by c_cl
def y(i):
    return (a_v*i_0/(n*F*D) * np.exp(alpha*z*F/RT*(E[i]-E_eq)))

## concentration with rxn plugged in
def c_cl(i):
    g = np.sqrt(y(i))
    b = beta(i)
    ret = []
    for z in x:
        a = 1.1*np.exp(-2*b) + ((c_feed*RT-(1.1*np.exp(-2*b)))/
                                (np.exp(-2*L*g)+1))*(np.exp(g*(z-2*L))+np.exp(-g*z))
        ret = np.append(ret,a)
    return ret
    
def current(i):
    g = np.sqrt(y(i))
    b = beta(i)
    ret = []
    z = 0;
    dcdx = g*((c_feed*RT-(1.1*np.exp(-2*b)))/
              (np.exp(-2*L*g)+1))*(np.exp(g*(z-2*L))-np.exp(-g*z))
    ## flux in here = 2*j_e from Yoon ?
    j_in = -D*dcdx
    
    return n*F*j_in
```

```python
plt.figure(figsize=(6,6))
for i in range(len(E)):
    plt.plot(x/L, c_cl(i), label=(str(round(E[i],2)) + 'V'))

plt.title('Concentration of Hydrogen in CL over distance from GDL-CL interface')
plt.ylabel('Concentration (mol/m^3)')
plt.xlabel('z (x/L)')
##plt.yscale('log')
plt.grid(True,which="both", linestyle='--')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(6,6))
I = []
for i in range(len(E)):
    ## area = trapz(c_cl(i), dx=0.5e-5)
    I = np.append(I,current(i))
    
plt.plot(E, I)
plt.title('Total Current Generated vs Potential Applied')
plt.ylabel('Current Generated (A)')
plt.xlabel('Potential (V)')
##plt.yscale('log')
##plt.grid(True,which="both", linestyle='--')
plt.show()
```

# film-agglomerate spherical model
based on yoon 2011 (https://sci-hub.se/10.1149/1.3597644)

```python
R_a = 0.1e-6
del_f = 20e-9
R_sph = R_a+del_f

De = 4.793e-6
#dry ionomer: Df = 104e-8
Df = 104e-8
i_0 = 2e-4

E = 0.3
eps = 0.3
lmbda = 0.8
L_ca = 0.2
A_ca = 40
l_CL = 15e-6

## reference c*_oxidation and c*_reduction
c_o = 1
c_r = 1
```

```python
theta = i_0/(1-eps)*L_ca*A_ca/(l_CL*c_r*F)*np.exp(alpha*z*F*(E-E_eq)/RT)
phi = np.sqrt(theta*R_sph/De)
omega = De/Df * lmbda/np.sinh(phi*R_a/R_sph) * (-(R_a/R_sph)*np.sinh(phi*R_a/R_sph) + 
                                      (R_a/R_sph)**2*phi*np.cosh(phi*R_a/R_sph))
c_film_0 = c_feed/(R_sph/R_a*omega - omega + 1)

## concentration in agglomerate
def c_agg(zed):
    return (1/zed) * lmbda*c_film_0 * R_a/R_sph * np.sinh(phi*zed) / np.sinh(phi*R_a/R_sph)

## concentration in film
def c_film(zed):
    return c_feed + omega*c_film_0*(1-1/zed)
```

```python
plt.figure(figsize=(7,7))

zed_i = np.arange(0.05e-8, R_a, 0.05e-8) / R_sph
zed_o = np.arange(R_a, R_sph+0.1*del_f, 0.1*del_f) / R_sph

plt.plot(zed_i, c_agg(zed_i), label='Agglomerate')
plt.plot(zed_o, c_film(zed_o), label='Film')

plt.title('Concentration of hydrogen across agglomerate-film length scale')
plt.ylabel('Concentration (mol/m^3)')
plt.xlabel('zed (r/(R_a+del_f))')
##plt.yscale('log')
plt.grid(True,which="both", linestyle='--')
plt.legend()
plt.show()
```

# Combined model

$$
A \left[ -D \frac{\partial c_{bulk}}{\partial x} \mid _{x} - \left( -D \frac{\partial c_{bulk}}{\partial x} \right) \mid _{x+dx} \right] = \left( \frac{\text{rxn rate}}{\text{unit CL vol}} \right) A dx
$$

$$
-D \frac{\partial^2 c_{bulk}}{\partial x^2} = \frac{\text{reaction rate}}{\text{unit CL vol}}
$$

$$ \phi _{CL} = \frac{\text{void vol}}{\text{total CL volume}} $$

$$ \Delta V_{CL} (1-\phi_{CL}) = \text{solid vol in $\Delta V$ CL vol} $$

$$ \text{Num. of aggl. } = \frac{\Delta V_{CL} (1-\phi_{CL})}{\frac{4}{3} \pi R_{agg}^3} $$

$$ \text{Consum. per aggl. } = 4 \pi R_{agg}^2 \left[ -D_{agg} \frac{\partial c_{agg}}{\partial r} \right] $$

$$ \frac{\text{reaction rate}}{\text{unit CL vol}} = \left( \frac{\text{# aggl.}}{\text{unit vol}} \right) \left( \frac{\text{consump.}}{\text{aggl.}} \right) $$

$$ -D_{agg} \frac{\partial^2 c_{bulk}}{\partial r^2} = \left( \frac{1-\phi_{CL}}{\frac{4}{3} R_{agg}^3 } \right) \left( 4 \pi R_{agg}^3 \left[ -D_{agg} \frac{\partial c_{agg}}{\partial r} \right] \right) $$

```python
# D_CL: https://escholarship.org/content/qt7kr960nf/qt7kr960nf.pdf
# De: https://sci-hub.se/10.1016/j.ijhydene.2018.07.035
# D_H2mix = 4.9e-5 for r_pore ~ 200nm
D_CL = 4.793e-6
p = (1-phi)*De/(4/3*R_a*D_CL*np.sinh(phi*R_a/R_sph)
               ) * (R_sph/R_a) * ( phi*np.cosh(R_a/R_sph) - (R_sph/R_a)*np.sinh(phi*R_a/R_sph)
                                 ) * 1/(R_sph/R_a*omega - omega + 1)
rt_p = np.sqrt(p)
k1 = 1/(1+np.exp(-2*rt_p*L))

def combined_bulk(x):
    return c_feed*((1-k1)*np.exp(rt_p*x) + k1*np.exp(-rt_p*x))

plt.figure(figsize=(7,7))
x = np.arange(0, L+0.5e-6, 0.5e-6)
plt.plot(x, combined_bulk(x)/c_feed)

plt.title('Concentration of H2 in anode along CL thickness')
plt.ylabel('Ratio of concentration over inlet conc.')
plt.xlabel('Length along CL thickness')
#plt.yscale('log')
plt.grid(True,which="both", linestyle='--')
plt.show()
```
